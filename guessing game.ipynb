**GUESSING GAME IN PYTHON**

This is a guessing game.The device automatically generates a random number.Now the user will be asked to input a number. If the number entered by the user matches the number generated by the device then the user will win the game. 6 chances will be given to the user for guessing.If after 6 chances the user is not able to guess correctly then the user will loose the game.

```python
import random
lower=int(input("Enter the lower bound:- "))
upper=int(input("Enter the upper bound :- "))
y=int('7')
x=random.randint(lower,upper)
print("\n\tYou have only",y,"chances to guess the number")
count=0
while count<(y):
    count=count+1
    guess=int(int(input("Guess a number")))
    if (guess==x):
        print("Congratulations you won the game")
        break
    elif guess<x:
        print("Your guess is too small!")
    elif guess>x:
        print("Your guess is too high!")
if count>=(y):
    print("\nThe number is",x)
    print("\tBetter luck next time!")
```
